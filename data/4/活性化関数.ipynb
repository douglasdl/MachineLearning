```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
# Sigmoid
# 正規分布の合計値に形が似ている
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

x = np.arange(-5, 5, 0.1)
e = math.e

y = sigmoid(x)

plt.plot(x, y)
plt.show()
```

```python
# (参考) 正規分布とシグモイド関数を微分したグラフの比較
x = np.arange(-5, 5, 0.1)
a = (np.exp(-x**2/2)) / math.sqrt(2*math.pi)

# シグモイドを微分
y = (1-sigmoid(x))*sigmoid(x)

plt.plot(x,a,color="blue")
plt.plot(x,y,color="red")

plt.show()
```

```python
# tanh(ハイパボリックタンジェント)
# シグモイド関数に似ているが0以下の値がある
y = np.tanh(x)
plt.plot(x, y)
plt.show()
```

```python
#ソフトマックス関数
# 合計したら1で、途中から急速な勾配がついている
x = np.arange(-5, 5, 0.1)
a = np.exp(x)
sum = np.sum(a)
y = a / sum
plt.plot(x, y)
plt.show()
```

```python
#RELU & Leakey ReLU
x = np.arange(-5.0, 5.0, 0.1)
y1 = np.where(x > 0, x , 0.01 * x)
y = np.maximum(0, x)
plt.plot(x, y)
plt.plot(x, y1)
plt.show()
```

```python
#ReLUとLeakey ReLUの拡大
y1 = np.where(x > 0, x , 0.01 * x)
y = np.maximum(0, x)
plt.plot(x, y)
plt.plot(x, y1)
plt.ylim(-0.1, 0.1)
plt.show()
```

```python
# ステップ関数
# Yes / No判定をする出力段に利用される
x = np.arange(-5.0, 5.0, 0.1)
y = x > 0
plt.plot(x, y)
plt.ylim(-0.1, 1.1)
plt.show()
```
